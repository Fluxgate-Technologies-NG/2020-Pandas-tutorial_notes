```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline

from matplotlib.ticker import MaxNLocator
```

```python
np.random.seed(0)
apple_stock = np.random.randn(5)
ms_stock = np.random.randn(5)
#Date = pd.Series(pd.date_range('2015', periods=5, freq='Y'))
#idx = ['Year 1', 'Year 2', 'Year 3', 'Year 4', 'Year 5']
#Date.index = idx

Date = [2010,2011,2012,2013,2014]
```

```python
stocks = plt.plot(Date, apple_stock,'g',
        Date, ms_stock, 'r-.')

plt.xlabel('year')
plt.ylabel('Apple stock')
#plt.axis([2009, 2015, -1.5, 2.5])

plt.show()
```

```python
fig_1 = plt.figure(1, figsize=(10, 5), dpi=100)

apple_price = fig_1.add_subplot(121) #(x,y,z) shows rows, columns and position
ms_price = fig_1.add_subplot(122)

apple_price.plot(Date, apple_stock)
ms_price.plot(Date, ms_stock, 'r--')

plt.show()
```

```python
#Alternatively
fig, axes = plt.subplots(1,2, figsize=(10,5), dpi=100)
axes[0].plot(Date, apple_stock)
axes[1].plot(Date, ms_stock, 'r--')
plt.show()
```

```python
#adding more than two subplots, make sure to reference the axes[rows,col]
fig, axes = plt.subplots(2,2, figsize=(10,5), dpi=100)
axes[0,0].plot(Date, apple_stock)
axes[1,0].plot(Date, ms_stock, 'r--')
plt.show()
```
