```python
#Are you scraping a webpage using read_html(), but it returns too many tables? 😕

#Use the 'match' parameter to find tables that contain a particular string! 🧶

url = 'https://en.wikipedia.org/wiki/List_of_most-followed_Instagram_accounts'

ig_page = pd.read_html(url)

len(ig_page)
    
```

```python
matching_tables = pd.read_html(url, match ='Profession')
```

```python
matching_tables[0].head()
```

```python
#Concatenante the two tables on IG i.e ig_page[0] and ig_page[1]

ig_long = pd.concat([ig_page[0], ig_page[1]])
ig_long.head()
```

```python
ig_long
```

```python
ig_long['Brand account'].fillna('Yes', inplace=True)
ig_long
```

```python
ig_long.rename(columns={'Followers(millions)': 'Followers in millions'}).head()
```

```python
#To select only profession as footballer
ig_long.loc[ig_long['Profession']=='Footballer']
```

```python
#Want to select from a DataFrame by label *and* position?

#Most readable approach is to chain "loc" (selection by label) and "iloc" (selection by position).
import pandas as pd
countries = pd.read_csv('../Data/gapminder.tsv', sep='\t')
countries
```

```python
countries.rename(str.upper, axis =1).head()
```

```python
countries.iloc[0:10].loc[:, ['country','year', 'pop']]
```

```python
#Reverse column order in a DataFrame:
#df.loc[:, ::-1]

countries.loc[:, ::-1]

#Reverse row order:
#df.loc[::-1]
countries.loc[::-1]

#Reverse row order and reset the index:
#df.loc[::-1].reset_index(drop=True)
countries.loc[::-1].reset_index(drop=True)
```

```python
#Want to know the *count* of rows that match a condition?
#(condition).sum()

#Want to know the *percentage* of rows that match a condition?
#(condition).mean()

(countries.continent=='Africa').sum()
```

```python
(countries.gdpPercap<1000).sum()
```

```python
import numpy as np
(countries.continent=='Asia').agg(np.mean)
```

```python
(countries.lifeExp>50).agg(np.mean)
```
