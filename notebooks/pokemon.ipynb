DATA FILTERING

```python
#you use loc to satisfy conditions using & or use seperator 
df.loc[df['Name'].str.contains('Mega')].head(3)
```

```python
df.loc[~df['Name'].str.contains('Mega')].head(3)
#here ~ means not Name
```

```python
#we can also pass regex expression
import re
df.loc[df['Type 1'].str.contains('Fire|Grass', regex=True)].head()
#here| stands for or

```

```python
df.loc[df['Type 1'].str.contains('fire|grass', flags=re.I, regex=True)].head()
#the flags.re.I ignores the capitalization in the strings contained in the argument
```

```python
#Lets say we want to filter the name column with all names starting with ur
df.loc[df['Name'].str.contains('^ur[a-z]*', flags=re.I, regex=True)].head()
#note the *sign means Zero or more and ^ to make the charaters start with ur
```
